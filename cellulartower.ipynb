```python
import pandas as pd
import geopandas as gpd
```

```python
data = gpd.read_file("assets/celltowers.geojson")
data.head(5)
```

```python
wa_data = gpd.read_file("assets/wacounties.json")
wa_df_data = pd.DataFrame(wa_data);
wa_df_data
```

```python
# use filtered dataframe to find unique values
print(sorted(wa_data.LICENSEE.unique()))
```

```python
# wa_data needs to contain all the count of cell towers in each individual county
county_group = data.groupby("LOCCOUNTY", as_index=False).count()
county_group['count'] = county_group.FID
result = county_group[['LOCCOUNTY', 'count']]
result['LOCCOUNTY'] = result['LOCCOUNTY'].str.lower().str.capitalize()
result.head(3)

# rename columns for merge
result.columns =['NAME', 'count']
result.head(3)
```

```python
# merge two dataframes for export json file
final = wa_data.merge(result, on = "NAME")
```

```python
# convert back to json
with open('wacountydata.json', 'w') as f:
        f.write(final.to_json())
```
